#Data Cleaning

```python
df_cleaned = df.copy()
```

```python
df_cleaned.replace(' ?', np.nan, inplace=True)
```

```python
missing_table = pd.DataFrame({
    'Missing Values': df_cleaned.isnull().sum(),
    'Percentage (%)': (df_cleaned.isnull().sum() / len(df)) * 100
})

missing_table
```

```python
df_cleaned.columns = (
    df_cleaned.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)
```

```python
df_cleaned.head()
```

```python
df_cleaned = df_cleaned.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
cols = df_cleaned.select_dtypes(include='object').columns

for col in cols:
    print(f"{col}: {df_cleaned[col].unique()}")
    print('#'*50)
```

```python
df_cleaned = df_cleaned.rename(columns={
    "fnlwgt": "final_weight",
    "education_num": "years_of_education"
})
```

```python
df_cleaned.duplicated().sum()
```

```python
df_cleaned=df_cleaned.drop_duplicates()
```

```python
le = LabelEncoder()
df_cleaned["sex"] = le.fit_transform(df_cleaned["sex"])
df_cleaned["income"] = le.fit_transform(df_cleaned["income"])
```

```python
df_cleaned.head()
```

```python
df_cleaned.drop("final_weight",axis=1,inplace=True)
```

```python
df_cleaned.head()
```

```python
df_cleaned.info()
```

```python
for col in ["workclass", "occupation", "native_country"]:
    df_cleaned[col].fillna(df_cleaned[col].mode()[0], inplace=True)
```
